## Background Research:

**What is the optimal number of bins?**

- Sturges Rule: In most stasticial packages the optimal number of bins is created using this rule. It is a simple rule, essentially it is the following formula:

$K = 1 + 3. 322*log(N)$

where:

$K$ = number of class intervals (bins).

$N$ = number of observations in the set.

$log$ = logarithm of the number.

## Methodology:

1) Use Sturges rule to caclulate the idea number of bins.

2) Plot the histogram using sturges rule to find the ideal number of bins.

3) Normalize the data for the other data set.

4) Plot the normalized distribution for the other dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

### Read in the Data

```python
iris = pd.read_csv('data/iris.csv')
glass = pd.read_csv('data/glass.csv')
```

### Using Sturges rule create Histogram with ideal number of bins

```python
def sturges(N: str):
    """
    The purpose of this method is to return the ideal number of bins given the number of datapoints
    """
    K = 1 + (3.322 * math.log(N))
    return round(K)
```

```python
plt.figure(figsize=(10,10))
plt.title("Distribution of Refractive Index")
plt.hist(glass['refractive_index'], bins=sturges(len(glass['refractive_index'])), color='darkmagenta')
plt.xlabel('Refractive Index')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
